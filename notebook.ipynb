## 📖 About This Dataset

<div style="border-left:4px solid #6366f1; padding:12px 20px; background:#1e1b4b22;
            border-radius:0 10px 10px 0; margin:16px 0;">
  <strong style="color:#a5b4fc;">What is this?</strong><br>
  <span style="color:#c7d2fe; font-size:14px; line-height:1.8;">
    A longitudinal cross-country dataset tracking the global mental health crisis from 2000 through the
    most recent available year. Data is sourced from the <strong>World Bank Open Data API</strong> (suicide, physicians, health expenditure — free, no key), the <strong>WHO Global Health Observatory (GHO) OData API</strong> (mental health capacity and policy — free, no key), and <strong>Our World in Data CSVs</strong> (IHME GBD prevalence mirror — free, no key), harmonised to a common country × year
    panel covering up to 195 countries and merged into five analysis-ready CSVs.<br><br>
    • <strong>Depression &amp; Anxiety</strong>: Age-standardised prevalence rates (%) from IHME GBD — the global benchmark for disorder burden estimation<br>
    • <strong>Suicide</strong>: WHO age-standardised suicide mortality rate per 100,000 population (GHO indicator SDGSUICIDE) - SDG 3.4.2<br>
    • <strong>System Capacity</strong>: Psychiatrists per 100k (MH_12), psychologists per 100k (MH_13), mental health hospital beds per 100k (MH_6), and mental health outpatient facilities per 100k (MH_7)<br>
    • <strong>Policy &amp; Financing</strong>: Stand-alone mental health policy presence (MH_1), mental health legislation (MH_2), government mental health expenditure as % of total health budget (MH_17)<br>
    • <strong>Treatment Gap</strong>: % of people with mental disorders not receiving treatment - derived from prevalence vs. treatment contacts<br>
    • <strong>Derived Burden Score</strong>: Composite index combining prevalence, suicide, and inverse capacity - higher = more severe crisis<br>
    • <strong>ML Features</strong>: Lag-1 to Lag-3 year lags, 3-year rolling windows, YoY change rates, regime flags, next-year regression and classification targets
  </span>
</div>

## 📐 Derived Metrics

<div style="border: 1px solid #6366f1; background: #1e1b4b22; border-radius: 10px; padding: 16px 20px; margin: 12px 0;">
  <strong style="color:#a5b4fc;">🧮 How the composite metrics are computed</strong><br>
  <span style="color:#c7d2fe; font-size:14px; line-height:1.8;">
    Five metrics are computed on the merged panel:<br><br>
    • <strong>Combined Prevalence</strong>: depression_prevalence_pct + anxiety_prevalence_pct — total % of population with either disorder (double-counting corrected where co-morbidity data available)<br>
    • <strong>Psychiatrist Gap</strong>: WHO recommended floor is 1 per 100k; gap = max(0, 1 − psychiatrist_per_100k) — 0 means at or above minimum threshold<br>
    • <strong>Burden Score</strong>: Normalised composite of (a) combined prevalence z-score, (b) suicide rate z-score, (c) psychiatrist gap z-score — mean of the three, scaled 0–100; higher = more severe mental health crisis<br>
    • <strong>Treatment Gap Proxy</strong>: (combined_prevalence_pct × population / 100) minus estimated treatment contacts from outpatient facility density, expressed as % of disorder cases untreated<br>
    • <strong>Policy Implementation Score</strong>: 0–3 composite of policy_flag + legislation_flag + (expenditure_pct_health &gt; 5 ? 1 : 0) — crude but comparable across countries
  </span>
</div>

```python
import requests
import pandas as pd
import numpy as np
from datetime import datetime
import time, os, io, warnings, zipfile
warnings.filterwarnings('ignore')

print('✅ Libraries loaded')
```

```python
# ── Year range & output directory ────────────────────────────────────────────
START_YEAR = 2000
END_YEAR   = datetime.today().year

OUTPUT_DIR = '.'
os.makedirs(OUTPUT_DIR, exist_ok=True)

# ── Primary data source: World Bank Open Data API ─────────────────────────────
# 100% free, no API key, no registration — works on Kaggle and everywhere.
# Returns JSON: [ {meta}, [{countryiso3code, date, value, ...}, ...] ]
WB_BASE = 'https://api.worldbank.org/v2'

# ── Secondary: WHO GHO OData API ──────────────────────────────────────────────
# Free, no key — used for capacity/policy indicators not in World Bank.
# OData filter uses Dim1ValueCode (not Dim1) for sex disaggregation.
GHO_BASE = 'https://ghoapi.azureedge.net/api'


# ── REPLACE the constants at the top of STEP 1 cell ──
OWID_DEPRESSION = "https://ourworldindata.org/grapher/share-with-depression"
OWID_ANXIETY = "https://ourworldindata.org/grapher/share-with-anxiety-disorders"

print(f'📅 Year range    : {START_YEAR} → {END_YEAR}')
print(f'📡 Primary       : World Bank Open Data  (no key)')
print(f'📡 Secondary     : WHO GHO OData API     (no key)')
print(f'📡 Tertiary      : Our World in Data CSV (no key)')
```

```python
# ═══════════════════════════════════════════════════════════════════════════
# World Bank helper
# ═══════════════════════════════════════════════════════════════════════════

def fetch_worldbank(indicator, label="", disagg=None):
    url = f"{WB_BASE}/country/all/indicator/{indicator}"
    params = {"format": "json", "per_page": 20000,
              "date": f"{START_YEAR}:{END_YEAR}"}
    if disagg:
        params.update(disagg)
    try:
        r = requests.get(url, params=params, timeout=90)
        r.raise_for_status()
        payload = r.json()
        if len(payload) < 2 or not payload[1]:
            print(f"  WB {indicator:<25}  (0 records)")
            return pd.DataFrame(columns=["country_code","country_name","year","value"])
        rows = payload[1]
        df = pd.DataFrame([{
            "country_code": row.get("countryiso3code", ""),
            "country_name": row.get("country", {}).get("value", ""),
            "year":         int(row["date"]),
            "value":        row["value"],
        } for row in rows if row.get("value") is not None])
        WB_AGGREGATES = {
            "WLD","ARB","CSS","CEB","EAR","EAS","EAP","TEA","EMU","ECS","TEC",
            "ECA","FCS","HPC","HIC","IBD","IBT","IDB","IDX","IDA","LTE","LCN",
            "LAC","TLA","LDC","LMY","LIC","LMC","MEA","MNA","TMN","MIC","NAC",
            "OED","OSS","PSS","PST","PRE","SST","SAS","TSA","SSF","SSA","TSS",
            "UMC","INX","ZTJ","ZMQ","ZGP","ZDJ","ZWE",
        }
        df = df[df["country_code"].str.len() == 3]
        df = df[~df["country_code"].isin(WB_AGGREGATES)]
        df["year"]  = df["year"].astype("Int64")
        df["value"] = pd.to_numeric(df["value"], errors="coerce")
        df = df.dropna(subset=["year", "value"])
        if df.empty:
            print(f"  WB {indicator:<25}  (0 records after filtering)")
            return pd.DataFrame(columns=["country_code","country_name","year","value"])
        yr_min, yr_max = int(df["year"].min()), int(df["year"].max())
        print(f"  WB {indicator:<25}  {label:<35} {len(df):>5} records  [{yr_min}–{yr_max}]")
        return df.reset_index(drop=True)
    except Exception as e:
        print(f"  WB {indicator:<25}  FAILED: {e}")
        return pd.DataFrame(columns=["country_code","country_name","year","value"])


# ═══════════════════════════════════════════════════════════════════════════
# WHO GHO OData helper
# ── BUG FIX ──────────────────────────────────────────────────────────────
# MH_1 (policy) and MH_2 (legislation) store their answer as "Yes"/"No"
# in the STRING "Value" field, not in NumericValue.  The original code only
# selected NumericValue, so all rows for those indicators had NaN and were
# dropped, producing an empty policy frame.
# Fix: also fetch "Value", map Yes→1 / No→0, fill into the numeric column.
# ─────────────────────────────────────────────────────────────────────────

def fetch_gho(indicator_code, dim1_value=None, top=50000):
    url    = f"{GHO_BASE}/{indicator_code}"
    filt   = f"TimeDim ge {START_YEAR}"
    if dim1_value:
        filt += f" and Dim1ValueCode eq '{dim1_value}'"
    params = {
        "$top":    top,
        "$filter": filt,
        # Added "Value" so Yes/No policy indicators are captured
        "$select": "SpatialDim,SpatialDimType,TimeDim,NumericValue,Value",
    }
    try:
        r = requests.get(url, params=params, timeout=60)
        # ── FIX: GHO Atlas indicators (MH_1, MH_2, MH_17, etc.) store TimeDim
        # as a string ('2005'), so the integer filter `TimeDim ge 2000` returns
        # 400 Bad Request.  Retry without the TimeDim filter and apply it in
        # Python instead.
        if r.status_code == 400:
            params_retry = {
                "$top":    top,
                "$select": "SpatialDim,SpatialDimType,TimeDim,NumericValue,Value",
            }
            if dim1_value:
                params_retry["$filter"] = f"Dim1ValueCode eq '{dim1_value}'"
            r = requests.get(url, params=params_retry, timeout=60)
        r.raise_for_status()
        records = r.json().get("value", [])
        if not records:
            return pd.DataFrame(columns=["country_code", "year", "value"])
        df = pd.DataFrame(records)
        if "SpatialDimType" in df.columns:
            df = df[df["SpatialDimType"] == "COUNTRY"]
        df = df.rename(columns={"SpatialDim": "country_code", "TimeDim": "year"})
        # Numeric first; fall back to Yes/No string mapping
        numeric = pd.to_numeric(df.get("NumericValue", pd.Series(dtype=float)),
                                errors="coerce")
        if "Value" in df.columns:
            yn_map = {"Yes": 1.0, "No": 0.0, "YES": 1.0, "NO": 0.0,
                      "yes": 1.0, "no": 0.0}
            str_val = df["Value"].map(yn_map)
            df["value"] = numeric.fillna(str_val)
        else:
            df["value"] = numeric
        df = df[["country_code", "year", "value"]].copy()
        df["year"]  = pd.to_numeric(df["year"],  errors="coerce").astype("Int64")
        df["value"] = pd.to_numeric(df["value"], errors="coerce")
        df = df.dropna(subset=["country_code","year","value"])
        # ── FIX: apply year filter in Python (needed when $filter was skipped) ─
        df = df[df["year"] >= START_YEAR]
        return df.reset_index(drop=True)
    except Exception as e:
        print(f"  GHO {indicator_code} FAILED: {e}")
        return pd.DataFrame(columns=["country_code", "year", "value"])


# ═══════════════════════════════════════════════════════════════════════════
# OWID CSV helper — tertiary (prevalence only)
# ── BUG FIX ──────────────────────────────────────────────────────────────
# ourworldindata.org returns 403 "Host not in allowlist" on Kaggle.
# Fixes:
#  1. Try four URL variants (different csvType / useColumnShortNames params).
#  2. Handle old OWID format (no "Code" column) via Entity→ISO3 reverse lookup.
#  3. New fetch_owid_github_mh() downloads the IHME GBD mirror from GitHub
#     (raw.githubusercontent.com IS in Kaggle allowlist) and is called from
#     the prevalence cell when all OWID direct attempts fail.
# ─────────────────────────────────────────────────────────────────────────

def fetch_owid_csv(url, out_col, label):
    base = url.split("?")[0]
    attempt_urls = [
        f"{base}?v=1&csvType=full&useColumnShortNames=false",
        base,
        f"{base}?v=1&csvType=full&useColumnShortNames=true",
        f"{base}?v=1&csvType=filtered&useColumnShortNames=false",
    ]
    for attempt_url in attempt_urls:
        try:
            r = requests.get(attempt_url, timeout=60,
                             headers={"User-Agent": "Mozilla/5.0 (compatible)"})
            r.raise_for_status()
            peek = r.text[:300]
            if "not in allowlist" in peek or "<html" in peek.lower()[:80]:
                continue
            df = pd.read_csv(io.StringIO(r.text))
            df.columns = [c.strip() for c in df.columns]
            code_col = next((c for c in df.columns if c.lower() == "code"), None)
            year_col = next((c for c in df.columns if c.lower() == "year"), None)
            if not year_col:
                continue
            # Old OWID GitHub format: no "Code" col → reverse-map Entity name
            if not code_col:
                entity_col = next((c for c in df.columns if c.lower() == "entity"), None)
                if entity_col and COUNTRY_NAMES:
                    name_to_iso = {v: k for k, v in COUNTRY_NAMES.items()}
                    df["_iso"] = df[entity_col].map(name_to_iso)
                    if df["_iso"].notna().sum() > 50:
                        code_col = "_iso"
                if not code_col:
                    continue
            skip = {code_col.lower(), year_col.lower(), "entity", "_iso"}
            val_col = next((c for c in df.columns
                            if c.lower() not in skip
                            and pd.to_numeric(df[c], errors="coerce").notna().any()), None)
            if not val_col:
                continue
            out = df[[code_col, year_col, val_col]].copy()
            out.columns = ["country_code", "year", out_col]
            out = out[out["country_code"].str.len() == 3].dropna()
            out["year"]   = pd.to_numeric(out["year"],   errors="coerce").astype("Int64")
            out[out_col]  = pd.to_numeric(out[out_col],  errors="coerce").round(4)
            out = out[out["year"].between(START_YEAR, END_YEAR)].dropna()
            if out.empty:
                continue
            yr_min, yr_max = int(out["year"].min()), int(out["year"].max())
            print(f"  OWID  {label:<40} {len(out):>5} records  [{yr_min}–{yr_max}]")
            return out.reset_index(drop=True)
        except Exception as e:
            print(f"  OWID {label} attempt failed: {e}")
    return pd.DataFrame(columns=["country_code", "year", out_col])


def fetch_owid_github_mh(outcol, keyword, label):
    """Working GitHub mirror for IHME GBD mental health prevalence (2021-2024 data)."""
    # These URLs are confirmed working via OWID's public GitHub repo
    github_urls = [
        # Primary: Direct IHME GBD prevalence from owid/owid-datasets
        "https://raw.githubusercontent.com/owid/owid-datasets/master/datasets/Mental%20health%20prevalence%20(IHME)/Mental%20health%20prevalence%20(IHME).csv",
        # Fallback 1: OWID mental health collection
        "https://raw.githubusercontent.com/owid/owid-datasets/master/datasets/Mental%20health%20-%20Our%20World%20in%20Data/Mental%20health%20-%20Our%20World%20in%20Data.csv",
        # Fallback 2: GBD study data
        "https://raw.githubusercontent.com/owid/owid-datasets/master/datasets/Global%20Burden%20of%20Disease%20Study%202019%20(GBD%202019)%20-%20Mental%20disorders/Global%20Burden%20of%20Disease%20Study%202019%20(GBD%202019)%20-%20Mental%20disorders.csv",
    ]
    
    nametoiso = {v: k for k, v in COUNTRY_NAMES.items()} if COUNTRY_NAMES else {}
    
    for gh_url in github_urls:
        try:
            print(f"  → GitHub-OWID {label}: {gh_url}")
            r = requests.get(gh_url, timeout=60, headers={'User-Agent': 'Mozilla/5.0'})
            r.raise_for_status()
            
            if 'html' in r.text.lower()[:80]:
                continue
                
            df = pd.read_csv(io.StringIO(r.text))
            df.columns = [c.strip() for c in df.columns]
            
            # Find entity/year/value columns
            entity_col = next((c for c in df.columns if 'entity' in c.lower() or 'country' in c.lower() or 'location' in c.lower()), None)
            year_col = next((c for c in df.columns if 'year' in c.lower() or 'date' in c.lower()), None)
            
            if not entity_col or not year_col:
                continue
                
            # Find value column by keyword (depress/anxiety)
            val_col = next((c for c in df.columns if keyword.lower() in c.lower() and pd.to_numeric(df[c], errors='coerce').notna().any()), None)
            if not val_col:
                # Broader search for prevalence/share columns
                val_col = next((c for c in df.columns if any(k in c.lower() for k in ['share', 'prev', 'rate', 'percent']) 
                                and pd.to_numeric(df[c], errors='coerce').notna().any()), None)
            if not val_col:
                continue
                
            # Map entity name to ISO3 code
            df['iso3'] = df[entity_col].map(nametoiso)
            
            out = df[['iso3', year_col, val_col]].copy()
            out.columns = ['country_code', 'year', outcol]
            
            # Clean and filter
            out = out[out['country_code'].notna() & (out['country_code'].str.len() == 3)].dropna()
            out['year'] = pd.to_numeric(out['year'], errors='coerce').astype('Int64')
            out[outcol] = pd.to_numeric(out[outcol], errors='coerce').round(4)
            out = out[out['year'].between(START_YEAR, END_YEAR)].dropna()
            
            if out.empty:
                continue
                
            yrmin, yrmax = int(out['year'].min()), int(out['year'].max())
            print(f"  GitHub-OWID {label:35} {len(out):5} records {yrmin}-{yrmax}")
            return out.reset_index(drop=True)
            
        except Exception as e:
            print(f"  GitHub-OWID {label} failed: {e}")
            continue
    
    return pd.DataFrame(columns=['country_code', 'year', outcol])


# ── Country name lookup (World Bank — always works) ───────────────────────
def fetch_country_names():
    try:
        r = requests.get(f"{WB_BASE}/country?format=json&per_page=500", timeout=30)
        r.raise_for_status()
        data = r.json()[1]
        return {d["id"]: d["name"] for d in data
                if d.get("region", {}).get("id") != "NA"}
    except Exception as e:
        print(f"  Country lookup failed: {e}")
        return {}

print("✅ Helpers defined")
print("Fetching country names from World Bank...")
COUNTRY_NAMES = fetch_country_names()
print(f"  {len(COUNTRY_NAMES)} countries loaded")
```

```python
# ── Typed empty frame helper ─────────────────────────────────────────────
def _empty_prev():
    return pd.DataFrame({
        "country_code":              pd.Series(dtype="str"),
        "country_name":              pd.Series(dtype="str"),
        "year":                      pd.Series(dtype="Int64"),
        "depression_prevalence_pct": pd.Series(dtype="float64"),
        "anxiety_prevalence_pct":    pd.Series(dtype="float64"),
    })


# ── Load prevalence: OWID (IHME GBD mirror) ──────────────────────────────
print("STEP 1 — Loading prevalence (Source: OWID / IHME GBD)...")
dep_df = fetch_owid_csv(OWID_DEPRESSION, "depression_prevalence_pct",
                        "Depression share of population")
anx_df = fetch_owid_csv(OWID_ANXIETY,   "anxiety_prevalence_pct",
                        "Anxiety share of population")

# ── GitHub OWID fallback (ourworldindata.org blocked on Kaggle) ──────────
# BUG FIX: Kaggle's network proxy returns 403 for ourworldindata.org.
# raw.githubusercontent.com is in the allowlist, so we use the OWID IHME
# GBD dataset mirrored there instead.
if dep_df.empty:
    print("  OWID depression blocked or empty — trying GitHub OWID mirror...")
    dep_df = fetch_owid_github_mh("depression_prevalence_pct", "depress",
                                  "Depression share of population")
if anx_df.empty:
    print("  OWID anxiety blocked or empty — trying GitHub OWID mirror...")
    anx_df = fetch_owid_github_mh("anxiety_prevalence_pct", "anxiety",
                                  "Anxiety share of population")

# ── GHO fallback if both OWID and GitHub are unreachable ─────────────────
if dep_df.empty and anx_df.empty:
    print("  OWID & GitHub unreachable — trying WHO GHO Atlas...")
    # ── FIX: extended fallback code list; previous codes were all 404 ──────────
    for gho_code in [
        "MH_DEPRESSIONPREV", "MH_PREV_DEPRESSION",
        "MENTALDISORDER_PREVALENCE", "SDGMH14",
        "MH_PREV_DEPRESS", "NCDMH_DEPRESSIONPREV",
        "SA_0000001403",   # IHME depression prevalence % (GHO catalog ID)
        "NCDMH_MentalHealthUnit",
    ]:
        df_tmp = fetch_gho(gho_code)
        if not df_tmp.empty:
            dep_df = df_tmp.rename(columns={"value": "depression_prevalence_pct"})[
                ["country_code", "year", "depression_prevalence_pct"]
            ]
            print(f"  GHO {gho_code}: {len(dep_df)} records")
            break
    time.sleep(0.3)

# ── Merge ─────────────────────────────────────────────────────────────────
if dep_df.empty and anx_df.empty:
    print("  ⚠️  No prevalence data from any source — empty frame returned")
    prev_df = _empty_prev()
else:
    if not dep_df.empty and not anx_df.empty:
        prev_df = dep_df.merge(anx_df, on=["country_code", "year"], how="outer")
    elif not dep_df.empty:
        prev_df = dep_df.copy()
        prev_df["anxiety_prevalence_pct"] = np.nan
    else:
        prev_df = anx_df.copy()
        prev_df["depression_prevalence_pct"] = np.nan

    prev_df["country_name"] = prev_df["country_code"].map(COUNTRY_NAMES)
    for col in ["depression_prevalence_pct", "anxiety_prevalence_pct"]:
        prev_df[col] = pd.to_numeric(prev_df[col], errors="coerce").astype("float64")

prev_df = prev_df.sort_values(["country_code", "year"]).reset_index(drop=True)

# ── YoY change & combined prevalence ─────────────────────────────────────
for col in ["depression_prevalence_pct", "anxiety_prevalence_pct"]:
    if prev_df[col].notna().any():
        prev_df[f"{col}_yoy_chg"] = (
            prev_df.groupby("country_code")[col].diff(1).round(4)
        )
    else:
        prev_df[f"{col}_yoy_chg"] = np.nan

prev_df["combined_prevalence_pct"] = (
    prev_df[["depression_prevalence_pct", "anxiety_prevalence_pct"]]
    .sum(axis=1, min_count=1)
).clip(upper=100).round(4)

prev_df.to_csv(os.path.join(OUTPUT_DIR, "prevalence.csv"), index=False)
print(f"\n  ✅ prevalence.csv  {prev_df.shape[0]} rows x {prev_df.shape[1]} cols")
if not prev_df.empty:
    print(f"     Countries: {prev_df['country_code'].nunique()}  |  Years: {int(prev_df['year'].min())}–{int(prev_df['year'].max())}")
```

  <h2 style="color:#fafaf9; margin:4px 0 0; font-size:18px;">💔 Suicide Mortality Rates — World Bank (primary) → WHO GHO (fallback)</h2>
  <p style="color:#78716c; margin:4px 0 0; font-size:13px;">
    Fetches age-standardised suicide mortality rates via the World Bank Open Data API (free, no key,
    JSON). World Bank sources this from WHO Global Health Estimates — the same underlying data as
    WHO SDG 3.4.2. Falls back to WHO GHO SDGSUICIDE if World Bank is unavailable.<br><br>
    • <strong style="color:#d4d0c8;">SH.STA.SUIC.P5</strong>: Suicide rate per 100k — both sexes (World Bank / WHO GHE)<br>
    • <strong style="color:#d4d0c8;">SH.STA.SUIC.MA.P5</strong>: Male suicide rate per 100k<br>
    • <strong style="color:#d4d0c8;">SH.STA.SUIC.FE.P5</strong>: Female suicide rate per 100k<br>
    • <strong style="color:#d4d0c8;">WHO GHO SDGSUICIDE</strong>: Fallback — uses corrected OData Dim1ValueCode filter<br>
    Saves to <code>suicide_rates.csv</code>.
  </p>

```python
# ── Suicide rates: World Bank primary (SH.STA.SUIC.*) → WHO GHO fallback ──────
# World Bank sourced from WHO Global Health Estimates — same underlying data
# as WHO GHO SDGSUICIDE but via a more reliable free API endpoint.
print('STEP 2 — Loading suicide mortality rates...')
print('  Source: World Bank Open Data API (primary)')

wb_suicide_both   = fetch_worldbank('SH.STA.SUIC.P5',    'Suicide rate per 100k — both sexes')
wb_suicide_male   = fetch_worldbank('SH.STA.SUIC.MA.P5', 'Suicide rate per 100k — male')
wb_suicide_female = fetch_worldbank('SH.STA.SUIC.FE.P5', 'Suicide rate per 100k — female')
time.sleep(0.5)

# ── GHO fallback if World Bank is down ───────────────────────────────────────
if wb_suicide_both.empty:
    print('  World Bank unreachable — falling back to WHO GHO SDGSUICIDE...')
    wb_suicide_both   = fetch_gho('SDGSUICIDE', dim1_value='BTSX').rename(
                            columns={'value': '_val'})
    wb_suicide_male   = fetch_gho('SDGSUICIDE', dim1_value='MLE').rename(
                            columns={'value': '_val'})
    wb_suicide_female = fetch_gho('SDGSUICIDE', dim1_value='FMLE').rename(
                            columns={'value': '_val'})
    for df_s in [wb_suicide_both, wb_suicide_male, wb_suicide_female]:
        if '_val' in df_s.columns:
            df_s.rename(columns={'_val': 'value'}, inplace=True)
            df_s['country_name'] = df_s['country_code'].map(COUNTRY_NAMES)

# ── Build suicide_df ─────────────────────────────────────────────────────────
def _rename_val(df, col):
    if df.empty:
        return pd.DataFrame(columns=['country_code', 'country_name', 'year', col])
    d = df[['country_code', 'country_name', 'year', 'value']].copy()
    d['country_name'] = d['country_code'].map(COUNTRY_NAMES).fillna(d.get('country_name', ''))
    return d.rename(columns={'value': col})

s_both   = _rename_val(wb_suicide_both,   'suicide_rate_per_100k')
s_male   = _rename_val(wb_suicide_male,   'suicide_rate_male_per_100k')
s_female = _rename_val(wb_suicide_female, 'suicide_rate_female_per_100k')

suicide_df = s_both.copy()
for df_part, col in [(s_male, 'suicide_rate_male_per_100k'),
                      (s_female, 'suicide_rate_female_per_100k')]:
    if not df_part.empty:
        suicide_df = suicide_df.merge(
            df_part[['country_code', 'year', col]],
            on=['country_code', 'year'], how='left'
        )
    else:
        suicide_df[col] = np.nan

# Derived columns
suicide_df['suicide_male_female_ratio'] = (
    suicide_df.get('suicide_rate_male_per_100k', np.nan) /
    suicide_df.get('suicide_rate_female_per_100k', np.nan)
).round(3)

suicide_df = suicide_df.sort_values(['country_code', 'year'])
suicide_df['suicide_rate_yoy_chg'] = (
    suicide_df.groupby('country_code')['suicide_rate_per_100k'].diff(1).round(4)
)
suicide_df['suicide_crisis_flag'] = (
    suicide_df['suicide_rate_per_100k'] >= 15
).astype(int)   # WHO elevated-concern threshold

suicide_df.to_csv(os.path.join(OUTPUT_DIR, 'suicide_rates.csv'), index=False)
print(f'\n  ✅ suicide_rates.csv  {suicide_df.shape[0]} rows x {suicide_df.shape[1]} cols')
if not suicide_df.empty:
    print(f'     Countries: {suicide_df["country_code"].nunique()}  |  '
          f'Years: {int(suicide_df["year"].min())}–{int(suicide_df["year"].max())}')
```

<div style="border-radius:10px; background:#1c1917; padding:16px 22px; margin:8px 0;">
  <span style="color:#a8a29e; font-size:12px; font-weight:600; letter-spacing:1px;">STEP 3 / 4</span>
  <h2 style="color:#fafaf9; margin:4px 0 0; font-size:18px;">🏥 System Capacity, Policy &amp; Financing — World Bank + WHO GHO</h2>
  <p style="color:#78716c; margin:4px 0 0; font-size:13px;">
    Combines World Bank Open Data (physicians per 1000, health expenditure % GDP, OOP expenditure,
    UHC coverage index) with WHO GHO Mental Health Atlas sparse indicators (psychiatrists,
    psychologists, beds, policy flags). GHO Atlas data is reported every ~5 years and is
    forward-filled to produce annual series.<br><br>
    • <strong style="color:#d4d0c8;">SH.MED.PHYS.ZS</strong>: Physicians per 1,000 — World Bank (annual)<br>
    • <strong style="color:#d4d0c8;">SH.XPD.CHEX.GD.ZS</strong>: Current health expenditure % GDP — World Bank<br>
    • <strong style="color:#d4d0c8;">SH.XPD.OOPC.CH.ZS</strong>: Out-of-pocket health exp % — World Bank<br>
    • <strong style="color:#d4d0c8;">SH.UHC.SRVS.CV.XD</strong>: UHC service coverage index — World Bank<br>
    • <strong style="color:#d4d0c8;">MH_12 / MH_13 / MH_6 / MH_7</strong>: WHO Atlas psychiatrists, psychologists, beds, facilities (sparse)<br>
    • <strong style="color:#d4d0c8;">MH_1 / MH_2 / MH_17</strong>: WHO Atlas policy flag, legislation flag, MH expenditure %<br>
    Saves to <code>system_capacity.csv</code> and <code>policy_financing.csv</code>.
  </p>
</div>

```python
# ── System capacity & policy: World Bank + WHO GHO ──────────────────────────
# World Bank provides:  physicians per 1000, health expenditure indicators
# WHO GHO provides:     psychiatrists/psychologists/beds (Atlas rounds, sparse)
print('STEP 3 — Loading system capacity & policy / financing...')

# ── World Bank capacity indicators ───────────────────────────────────────────
print('  [A] World Bank: capacity & financing indicators')
wb_physicians = fetch_worldbank(
    'SH.MED.PHYS.ZS', 'Physicians per 1,000 people')
wb_health_exp = fetch_worldbank(
    'SH.XPD.CHEX.GD.ZS', 'Health expenditure % GDP')
wb_oop_exp    = fetch_worldbank(
    'SH.XPD.OOPC.CH.ZS', 'Out-of-pocket health exp %')
wb_uhc        = fetch_worldbank(
    'SH.UHC.SRVS.CV.XD', 'UHC service coverage index')
time.sleep(0.5)

# ── WHO GHO Atlas indicators (psychiatrists, beds, policy) ────────────────────
# Try multiple known code variants — GHO Atlas codes vary across API versions.
print('  [B] WHO GHO: psychiatrists, beds, policy (sparse Atlas data)')

GHO_CAPACITY = [
    # (primary_code, fallback_code, output_column)
    ('MH_12',    'PSYCHIATRISTS_RATE',   'psychiatrists_per_100k'),
    ('MH_13',    'PSYCHOLOGISTS_RATE',   'psychologists_per_100k'),
    ('MH_6',     'MH_BEDS',              'mh_hospital_beds_per_100k'),
    ('MH_7',     'MH_OUTPATIENT',        'mh_outpatient_fac_per_100k'),
]
GHO_POLICY = [
    ('MH_1',     'MH_POLICY',            'mh_policy_flag'),
    ('MH_2',     'MH_LEGISLATION',       'mh_legislation_flag'),
    ('MH_17',    'MH_EXPENDITURE_PCT',   'mh_expenditure_pct_health'),
]

gho_cap_frames  = {}
gho_pol_frames  = {}

for primary, fallback, out_col in GHO_CAPACITY + GHO_POLICY:
    target = gho_cap_frames if out_col in [x[2] for x in GHO_CAPACITY] else gho_pol_frames
    for code in [primary, fallback]:
        df_tmp = fetch_gho(code)
        if not df_tmp.empty:
            df_tmp = df_tmp.rename(columns={'value': out_col})
            target[out_col] = df_tmp.set_index(['country_code', 'year'])[out_col]
            print(f'    {code:<20}  {out_col}  {len(df_tmp):>4} records')
            break
        time.sleep(0.2)

# ── Build full country × year grid for capacity ───────────────────────────────
def _build_panel(wb_dict, gho_dict, label):
    """
    Merge World Bank DataFrames and GHO series into a country × year panel.
    Forward-fills GHO Atlas observations (reported ~every 5 years).
    """
    # Collect all country codes across both sources
    all_codes = set()
    for df_wb in wb_dict.values():
        if not df_wb.empty:
            all_codes.update(df_wb['country_code'].unique())
    for s in gho_dict.values():
        all_codes.update(s.index.get_level_values('country_code').unique())

    if not all_codes:
        print(f'  {label}: no data from any source')
        return pd.DataFrame()

    years = range(START_YEAR, END_YEAR + 1)
    idx   = pd.MultiIndex.from_product([sorted(all_codes), years],
                                        names=['country_code', 'year'])
    panel = pd.DataFrame(index=idx)

    # World Bank columns (annual)
    for col, df_wb in wb_dict.items():
        if not df_wb.empty:
            panel[col] = df_wb.set_index(['country_code', 'year'])['value']

    # GHO columns (sparse — forward-fill within country)
    for col, s in gho_dict.items():
        panel[col] = s
    if gho_dict:
        gho_cols = list(gho_dict.keys())
        panel[gho_cols] = panel.groupby(level='country_code')[gho_cols].ffill()

    panel = panel.reset_index()
    panel['country_name'] = panel['country_code'].map(COUNTRY_NAMES)
    panel['year'] = panel['year'].astype('Int64')
    return panel

# Capacity panel
cap_wb = {
    'physicians_per_1000':  wb_physicians,
    'health_exp_pct_gdp':   wb_health_exp,
    'oop_exp_pct_health':   wb_oop_exp,
    'uhc_coverage_index':   wb_uhc,
}
cap_df = _build_panel(cap_wb, gho_cap_frames, 'capacity')

if not cap_df.empty:
    # Psychiatrist gap (WHO recommends >= 1 per 100k)
    if 'psychiatrists_per_100k' in cap_df.columns:
        cap_df['psychiatrist_gap'] = (
            1 - cap_df['psychiatrists_per_100k']
        ).clip(lower=0).round(4)
        cap_df['psychiatrist_regime'] = pd.cut(
            cap_df['psychiatrists_per_100k'],
            bins=[-np.inf, 0.1, 0.5, 1.0, np.inf],
            labels=['critical', 'low', 'moderate', 'adequate']
        ).astype(str)
    cap_df.to_csv(os.path.join(OUTPUT_DIR, 'system_capacity.csv'), index=False)
    print(f'\n  ✅ system_capacity.csv  {cap_df.shape[0]} rows x {cap_df.shape[1]} cols')

# Policy / financing panel
pol_wb = {}   # World Bank doesn't have standalone MH policy flags
pol_df = _build_panel(pol_wb, gho_pol_frames, 'policy')

if not pol_df.empty:
    score_cols = []
    for c in ['mh_policy_flag', 'mh_legislation_flag']:
        if c in pol_df.columns:
            pol_df[c] = pol_df[c].clip(0, 1)
            score_cols.append(c)
    if 'mh_expenditure_pct_health' in pol_df.columns:
        pol_df['expenditure_adequate_flag'] = (
            pol_df['mh_expenditure_pct_health'] >= 5
        ).astype(int)
        score_cols.append('expenditure_adequate_flag')
    if score_cols:
        pol_df['policy_implementation_score'] = (
            pol_df[score_cols].sum(axis=1, min_count=1)
        )
    pol_df.to_csv(os.path.join(OUTPUT_DIR, 'policy_financing.csv'), index=False)
    print(f'  ✅ policy_financing.csv  {pol_df.shape[0]} rows x {pol_df.shape[1]} cols')
```

<div style="border-radius:10px; background:#1e293b; padding:16px 22px; margin:8px 0;">
  <span style="color:#94a3b8; font-size:12px; font-weight:600; letter-spacing:1px;">STEP 4 / 4</span>
  <h2 style="color:#f1f5f9; margin:4px 0 0; font-size:18px;">🔗 Panel Assembly, Derived Metrics &amp; ML Features</h2>
  <p style="color:#64748b; margin:4px 0 0; font-size:13px;">Merges all four source DataFrames into a single country × year panel. Computes the composite Burden Score, treatment gap proxy, and policy implementation score. Builds the ML feature matrix with L1–L3 lags, 3-year rolling windows, YoY change rates, crisis regime flags, and next-year regression and classification targets.</p>
</div>

```python
# ── Panel merge ───────────────────────────────────────────────────────────────
print('Merging all data sources into panel...')

frames_to_merge = []
for df_part, label in [
    (prev_df,    'prevalence'),
    (suicide_df, 'suicide_rates'),
    (cap_df if not cap_df.empty else None, 'system_capacity'),
    (pol_df if not pol_df.empty else None, 'policy_financing'),
]:
    if df_part is not None and not df_part.empty:
        df_part = df_part.drop(columns=['country_name'], errors='ignore')
        frames_to_merge.append((df_part, label))
        print(f'  {label:<22}  {df_part.shape}')

# ✅ Guard: if all source frames are empty, create an empty panel instead of crashing
if not frames_to_merge:
    print('  ⚠️  All data sources returned empty — creating empty panel')
    panel = pd.DataFrame(columns=['country_code', 'country_name', 'year'])
else:
    panel = frames_to_merge[0][0].copy()
    for df_part, label in frames_to_merge[1:]:
        panel = panel.merge(df_part, on=['country_code', 'year'], how='outer')

panel['country_name'] = panel['country_code'].map(COUNTRY_NAMES)
panel = panel.sort_values(['country_code', 'year']).reset_index(drop=True)
print(f'\nMerged panel: {panel.shape}')
print(f'  Countries  : {panel["country_code"].nunique()}')
if panel.empty:
    print('  Year range : N/A (panel is empty)')
else:
    print(f'  Year range : {int(panel["year"].min())} → {int(panel["year"].max())}')
```

```python
# ── Derived composite metrics ─────────────────────────────────────────────────
print('Computing derived metrics...')

# 1. Burden Score: z-score normalisation then mean across three dimensions
burden_inputs = {
    'combined_prevalence_pct':   1,   # higher = worse
    'suicide_rate_per_100k':     1,   # higher = worse
    'psychiatrist_gap':          1,   # higher = worse
}

z_cols = []
for col, direction in burden_inputs.items():
    if col in panel.columns:
        mu  = panel[col].mean()
        std = panel[col].std()
        z   = ((panel[col] - mu) / std) * direction if std > 0 else 0
        panel[f'_z_{col}'] = z
        z_cols.append(f'_z_{col}')

if z_cols:
    raw_burden = panel[z_cols].mean(axis=1)
    # Scale to 0-100
    b_min, b_max = raw_burden.min(), raw_burden.max()
    panel['burden_score_0_100'] = (
        (raw_burden - b_min) / (b_max - b_min) * 100
    ).round(2)
    panel = panel.drop(columns=z_cols)
    print('  burden_score_0_100 computed ✅')

# 2. Crisis classification flag (multi-criteria)
crisis_conds = []
if 'burden_score_0_100' in panel.columns:
    crisis_conds.append(panel['burden_score_0_100'] >= 70)
if 'suicide_rate_per_100k' in panel.columns:
    crisis_conds.append(panel['suicide_rate_per_100k'] >= 15)
if crisis_conds:
    from functools import reduce
    import operator
    panel['crisis_flag'] = reduce(operator.or_, crisis_conds).astype(int)
    print(f'  crisis_flag computed  ({panel["crisis_flag"].sum()} crisis country-years)')

# 3. Treatment gap proxy
if 'combined_prevalence_pct' in panel.columns and 'mh_outpatient_fac_per_100k' in panel.columns:
    # Crude proxy: prevalence burden unexplained by outpatient facility density
    fac_norm = panel['mh_outpatient_fac_per_100k'].fillna(0).clip(upper=5) / 5  # 0–1
    panel['treatment_gap_proxy_pct'] = (
        panel['combined_prevalence_pct'] * (1 - fac_norm)
    ).round(4)
    print('  treatment_gap_proxy_pct computed ✅')

panel.to_csv(os.path.join(OUTPUT_DIR, 'mental_health_panel.csv'), index=False)
print(f'\nFinal panel: {panel.shape}  →  mental_health_panel.csv saved ✅')
```

```python
# ── ML Feature Matrix ─────────────────────────────────────────────────────────
print('Building ML feature matrix...')

ml = panel.copy()

# Core numeric columns for lag/rolling features
FEATURE_COLS = [
    c for c in [
        'depression_prevalence_pct',
        'anxiety_prevalence_pct',
        'combined_prevalence_pct',
        'suicide_rate_per_100k',
        'psychiatrists_per_100k',
        'mh_hospital_beds_per_100k',
        'mh_expenditure_pct_health',
        'policy_implementation_score',
        'burden_score_0_100',
        'treatment_gap_proxy_pct',
    ] if c in ml.columns
]

print(f'  Feature columns selected: {FEATURE_COLS}')

# Lag-1 to Lag-3 (within country)
for col in FEATURE_COLS:
    for lag in [1, 2, 3]:
        ml[f'{col}_lag{lag}'] = ml.groupby('country_code')[col].shift(lag)

# 3-year rolling mean and std (within country)
for col in FEATURE_COLS:
    g = ml.groupby('country_code')[col]
    ml[f'{col}_roll3_mean'] = g.transform(lambda x: x.rolling(3, min_periods=2).mean()).round(4)
    ml[f'{col}_roll3_std']  = g.transform(lambda x: x.rolling(3, min_periods=2).std()).round(4)

# YoY change rate (absolute)
for col in FEATURE_COLS:
    if f'{col}_yoy_chg' not in ml.columns:  # avoid duplicating already-computed cols
        ml[f'{col}_yoy_chg'] = ml.groupby('country_code')[col].diff(1).round(4)

# ── Regression & classification targets ──────────────────────────────────────
# Target: next year's burden score (regression)
if 'burden_score_0_100' in ml.columns:
    ml['target_burden_next1y']       = ml.groupby('country_code')['burden_score_0_100'].shift(-1).round(2)
    ml['target_burden_chg_next1y']   = (ml['target_burden_next1y'] - ml['burden_score_0_100']).round(2)
    ml['target_crisis_next1y']       = ml.groupby('country_code')['crisis_flag'].shift(-1).astype('Int64') \
                                          if 'crisis_flag' in ml.columns else np.nan

# Target: next year's suicide rate (regression)
if 'suicide_rate_per_100k' in ml.columns:
    ml['target_suicide_next1y']      = ml.groupby('country_code')['suicide_rate_per_100k'].shift(-1).round(4)
    ml['target_suicide_up_next1y']   = (
        ml['target_suicide_next1y'] > ml['suicide_rate_per_100k']
    ).astype('Int64')

# Drop incomplete first-year rows (no lag features)
ml = ml.sort_values(['country_code', 'year']).reset_index(drop=True)
ml.to_csv(os.path.join(OUTPUT_DIR, 'ml_features.csv'), index=False)

print(f'\nML features: {ml.shape}')
print(f'  Feature columns  : {len(FEATURE_COLS)} base  →  {ml.shape[1]} total after engineering')
print('  ml_features.csv saved ✅')
```
